```python
#step 1 encoding the dataset
import numpy as np

data = [
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
]

# Convert to numpy array
data = np.array(data, dtype=object)

# Split features and labels
X = data[:, :3].astype(float)
y_raw = data[:, 3]

# Label encoding
label_map = {
    'Apple': 0,
    'Banana': 1,
    'Orange': 2
}

y = np.array([label_map[label] for label in y_raw])

print("X:\n", X)
print("y:\n", y)
```

```python
#define euclidean_dist
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
#knn
class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict_one(self, x):
        distances = []

        # Step 1: compute distance to all training points
        for i in range(len(self.X_train)):
            dist = euclidean_distance(x, self.X_train[i])
            distances.append((dist, self.y_train[i]))

        # Step 2: sort by distance
        distances.sort(key=lambda t: t[0])

        # Step 3: pick k nearest
        k_nearest = distances[:self.k]

        # Step 4: voting
        labels = [label for i, label in k_nearest]
        counts = np.bincount(labels)

        return np.argmax(counts)
    def predict(self, X_test):
        predictions = []

        for x in X_test:
            pred = self.predict_one(x)
            predictions.append(pred)

        return np.array(predictions)
```

```python
#knn testing 
test_data = np.array([
    [118, 6.2, 0],  # Expected: Banana
    [160, 7.3, 1],  # Expected: Apple
    [185, 7.7, 2]   # Expected: Orange
])
knn=KNN(k=3)
knn.fit(X,y)
predictions = knn.predict(test_data)
print("Encoded predictions:", predictions)

#decoding back to fruits
reverse_map = {v: k for k, v in label_map.items()}
decoded = [reverse_map[p] for p in predictions]
print("Predicted fruits:", decoded)
```

```python
#evaluation
#predictions for different k

for k in [1,2,3,4,5]:
    knn = KNN(k=k)
    knn.fit(X, y)
    preds = knn.predict(test_data)
    decoded = [reverse_map[p] for p in preds]
    print(f"k={k} : {decoded}")


#low k : overfit
#high k: underfit
```
